```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.metrics import rand_score, adjusted_rand_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons
from sklearn.svm import SVC

from warnings import filterwarnings
filterwarnings("ignore")
```

```python
X, y = make_moons(n_samples=1000, noise=0.4, random_state=42)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.shape, X_test.shape
```

```python
for i in ['linear', 'poly', 'rbf', 'sigmoid']:
    svm_cls = SVC(kernel=i, random_state=42)
    svm_cls.fit(X_train, y_train)
    pred = svm_cls.predict(X_test)
    print(f"kernel={i}")
    print(f"adjusted_rand_score={adjusted_rand_score(labels_true=y_test, labels_pred=pred)}")
    print(f"rand_score={rand_score(labels_true=y_test, labels_pred=pred)}")
    print("-"*40)
```

```python
model = SVC(kernel="rbf", random_state=42)
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
print(classification_report(y_pred=pred, y_true=y_test))
```

```python
pred = model.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=pred)
plt.show()
```
